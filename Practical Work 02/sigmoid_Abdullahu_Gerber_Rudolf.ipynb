```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def sigmoid(z):
    return 1/(1+np.e**(-z))

def sigmoid_der(z):
    sig = np.diff(sigmoid(z))
    return sig

def log_sig_1st_der(z):
    return -1/z[1:]*sigmoid_der(z)

def log_sig_2nd_der(z):
    return np.diff(log_sig_1st_der(z))




```

```python
#a
z = np.arange(-10,10,0.1)
plt.subplot(1,1,1)
plt.title('a)')
plt.plot(sigmoid(z))
plt.grid()
```

```python
#b
plt.subplot(1,2,1)
plt.title('b)')
plt.plot(sigmoid_der(z))
plt.grid()

plt.subplot(1,2,2)
plt.plot(sigmoid(z)*(1-sigmoid(z)))
plt.grid()
```

```python
#c
plt.clf()
plt.subplot(1,2,1)
plt.title('c)')
plt.plot(log_sig_1st_der(z))
plt.grid()

plt.subplot(1,2,2)
plt.plot(log_sig_2nd_der(z))
plt.grid()
plt.show()
```

```python
x = np.arange(10**10,10**11,10**9)
plt.plot(log_sig_1st_der(x))
plt.title('c)')
plt.grid()
```

```python
#f
def c1(x):
    return (sigmoid(x)-1)**2

x = np.arange(-10,10,0.1)

plt.subplot(1,3,1)
plt.title('f)')
plt.plot(c1(x))
plt.grid()

plt.subplot(1,3,2)
plt.plot(np.diff(c1(x)))
plt.grid()

plt.subplot(1,3,3)
plt.plot(np.diff(np.diff(c1(x))))
plt.grid()




```

```python
def c2(x,w,y):
    return -(y*np.log(sigmoid(w*x))+(1-y)*np.log(1-sigmoid(x*w)))

x = np.arange(-10,10,0.1)
w = np.e**x
y = (x-np.min(x))/(np.max(x)-np.min(x))

plt.subplot(1,3,1)
plt.title('g)')
plt.plot(c2(x,w,y))
plt.grid()

plt.subplot(1,3,2)
plt.plot(np.diff(c2(x,w,y)))
plt.grid()

plt.subplot(1,3,3)
plt.plot(np.diff(np.diff(c2(x,w,y))))
plt.grid()
```
